#### 5.1.1.2 One Hot Encoding (Directo)
Hacemos One Hot Encoding a Tipo de Area, Ciudad y 'Tenant Preferred' (Inquilino preferido). Todas estas columnas no tienen muchas variables, lo que hace posible hacer este tipo feature engineering. Adicionalmente, son variables que no poseen ningun sistema jerárquico.

#### 5.1.1.3 Label Encoding

Realizamos el Label Encoding de la variable Point of Contact.
Se hace este tipo de transformación, ya que en el EDA se observó una notoria diferencia en la renta cuando el un agente inmobiliario es la persona contacto del inmueble.

Primero, agrupamos por la variable y medimos la renta media.
Luego, observamos su porcentaje en base al total.
Y por último, asignamos un valor que represente más o menos el salto existentes en las rentas medias, cuando se cambia a la persona contacto.

```python
#Agrupamos
F1_POC = F1.groupby('Point of Contact').agg({'Rent':np.median})
```

```python
#Calculamos proporciones en base al total.
F1_POC['Porcentaje2I'] = F1_POC['Rent'] / F1_POC['Rent'].sum()
```

```python
F1_POC['Porcentaje2I']*10
```

Sacamos el porcentaje del total de cada una de las medias y redondeamos.
De tal forma asignamos:

 - Numero 7 para Contact Agent
 - Numero 1 para Contact Builder
 - Nimero 2 para Contact Owner

 Ya que tenemos mayor probabilidad de tener precios mas altos si la vivienda es vendida por un agente inmobiliario.

```python
F1['Point of Contact'].replace({'Contact Agent': 7 , 'Contact Builder': 1, 'Contact Owner': 2}, inplace = True)
F1.sample(6)
```

Realizamos el Label Encoding de la variable 'Furnishing Status'. Se aplica la misma técnica aplicada con la variable 'Point of Contact'.

```python
F1_FS = F1.groupby('Furnishing Status').agg({'Rent':'median'})
F1_FS['Porcentaje'] = F1_FS['Rent'] / F1_FS['Rent'].sum()
```

```python
F1_FS['Porcentaje'] *10
```

Sacamos el porcentaje del total de cada una de las medias y redondeamos. Como esta en secuencia podemos asignar de tal forma: 

 - Numero 1 para Unfurnished
 - Numero 2 para Semi-Furnished
 - Nimero 4 para Furnished

 Ya que tenemos mayor probabilidad de tener precios mas altos si la vivienda es vendida por un agente inmobiliario.

```python
F1['Furnishing Status'].replace({'Unfurnished': 1 , 'Semi-Furnished': 2, 'Furnished': 4}, inplace = True)
F1.sample(6)
```

Utilizamos el metodo IQR

BS (Bigote SUPERIOR) = Q3 + IQR*1.5

BI (Bigote INFERIOR) = Q1 - IQR*1.5

IQR = RANGO INTERCUARTIL = Q3-Q1

```python
Q1 = hr_df['Rent'].quantile(q = 0.25)
Q3 = hr_df['Rent'].quantile(q = 0.75)

IQR = Q3-Q1

BS = Q3 + IQR*1.5
BI = Q1 - IQR*1.5

print('Q1:' + str(Q1))
print('Q3:' + str(Q3))
print()
print('IQR:' + str(IQR))
print()
print('BS:' + str(BS))
print('BI:' + str(BI))
```

##6.2Conclusiones CrossValidation y mejora de modelos de ML

Se utiliza el algoritmo de **Halving Grid Search** para buscar la mejor combinacion de hiperparametros, pero con el beneficio de reducir el tiempo de iteración.

**Nota: El CrossValidation utilizado por default es de 5.**

Los hiperparámetros elegidos por el algoritmo fueron:

1. 'criterion': 'poisson'
2. 'max_depth': None
3. 'max_features': 'log2'
4. 'n_estimators': 250


Con estos hiperparámetros el performance aumenta por las siguientes razones:

1. Aumenta el numero de **n_estimators**: El algoritmo eligio 250 vs los 100 que por default el algoritmo de RandomForestRegressor tiene. En la medida que aumentas la cantidad de arboles incentivas un mejor performance del modelo a costa de mayor tiempo de entrenamiento.
2. El **max_depth** utiliza el valor por defecto del algoritmo, por lo que las ramificaciones del algoritmo de random forest se expanden hasta que todas las hojas son puras o contienen menos de las muestras mínimas para la división. Al contener mayor numero de divisiones, se obtienen mayor número de ramificaciones de las cuales el modelo puede aprender. Por lo tanto, el modelo tiende a tener un mejor performance.
3. El **max_feature** hiperparámetro define el número total de features para cada arbol. El algoritmo deferminó que la mejor metodologia es la de la raiz cuadrada del total de features. 
4. El **criterion** es la metodologia utilizada por el algoritmo para determinar la calidad de cada ramificación. En este caso se optimiza con el **poisson**, el cual representa un pequeña mejora del criterio mse, el cual no solo toma en consideración que tan cerca se encuentra el valor del resultado, sino tambien las probabilidades en el momento de tomar una decision en un nodo específico.
